# McNemars test for the paired questions. Questions presented as text/tables written out vs questions presented as a screenshot

```python
# Import library
from statsmodels.stats.contingency_tables import mcnemar

# 2x2 table: rows = paired image (success/failure), cols = paired text (success/failure)
data = [[51, 9],
        [35, 39]]

print(mcnemar(data, exact=False, correction=False))
```

# Two-proportion z-test between accuracy of questions with only text and questions with only screenshot 

```python
from statsmodels.stats.proportion import proportions_ztest
# two-sided: unpaired text vs unpaired image
z, p = proportions_ztest([104, 59], [141, 157])
print(z)
print(p)
```

## one-sided two-proportion z-test between accuracy before knowledge cutoff and after knowledge cutoff

```python
from statsmodels.stats.proportion import proportions_ztest
# one-sided (pre > post), paired questions scored as image
z, p = proportions_ztest([187, 36], [378, 54], alternative='larger')
print('paired as image:', z, p)
# one-sided (pre > post), paired questions scored as text
z, p = proportions_ztest([212, 37], [378, 54], alternative='larger')
print('paired as text:', z, p)
```

# power

```python
import numpy as np
# power: minimum detectable effect at 80% power, alpha = 0.05, worst case p = 0.5
z = 1.96 + 0.84    # two-sided (text vs image, smaller group n = 141)
print(f"detectable effect size text vs image = {round(z*np.sqrt(0.5/141), 3)}")
z1 = 1.645 + 0.84  # one-sided (contamination check, smaller group n = 54)
print(f"detectable effect size contamination = {round(z1*np.sqrt(0.5/54), 3)}")
```
